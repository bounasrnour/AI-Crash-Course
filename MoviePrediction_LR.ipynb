##Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

##Reading CSV File

```python
data = pd.read_csv("/content/move_budget.csv")
data.head()
```

```python
data.describe()
```

```python
#production budget
X = pd.DataFrame(data, columns=['production_budget_usd'])
#revenue
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
print(X, "\n",y)
```

##Plotting

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title("Film Cost Vs Rrevenu")
plt.xlabel("Film Cost")
plt.xlabel("Gross Revenue")
plt.ylim(0, 3000000000)
plt.xlim(0, 415000000)
plt.show()
```

##Linear Regression

```python
linear_reg = LinearRegression()
linear_reg.fit(X,y)
linear_reg.coef_
linear_reg.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title("Film Cost Vs Rrevenu")
plt.xlabel("Film Cost")
plt.xlabel("Gross Revenue")
plt.ylim(0, 3000000000)
plt.xlim(0, 415000000)
plt.plot(X, linear_reg.predict(X), color='red', linewidth=3)
plt.show()
```

##Accuracy

```python
linear_reg.score(X, y)
```
